<h1>Using Autogradto calculate the deriavatives </h1>

<h3> Derivative for y=x^2</h3>

```python
import torch
x=torch.tensor(3.0,requires_grad=True)
y=x**2
y.backward()
print(x.grad)
```

<h3> For y=x^2, z=sin(y) ,dz/dx</h3>

```python
import torch
x=torch.tensor(3.0,requires_grad=True)
y=x**2
z=torch.sin(y)
z.backward()
#y.backward()
print(x.grad)
```

<h3> For a neural network along with back propagation(Derivative w.r.t to w and b) </h3>

```python
import torch
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))

x=torch.tensor(6.7)
y=torch.tensor(0.0)

w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)

#Starting with forward pass
z=(w*x)+b
y_pred=torch.sigmoid(z)
loss=binary_cross_entropy_loss(y_pred,y)

#backward propagation
loss.backward()
print(w.grad)
print(b.grad)
```
